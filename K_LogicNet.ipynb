In fixed learning rate, update step size is constant. May get stuck in local minima if ℓɼ is too large.

In decaying learning rate, it starts with larger steps to move quickly towards minimum. Decreases gradually to allow finer adjustments towards optimum.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

y_pred = np.linspace(0.01, 0.99, 100)

# Standard BCE
bce_1 = -np.log(y_pred)
bce_0 = -np.log(1 - y_pred)

# Weighted BCE (weight pos=3, neg=1)
wbce_1 = -3 * np.log(y_pred)
wbce_0 = -1 * np.log(1 - y_pred)

plt.figure(figsize=(10,6))
plt.plot(y_pred, bce_1, label='BCE y=1', color='green')
plt.plot(y_pred, bce_0, label='BCE y=0', color='blue')
plt.plot(y_pred, wbce_1, '--', label='WBCE y=1 (weight=3)', color='green')
plt.plot(y_pred, wbce_0, '--', label='WBCE y=0 (weight=1)', color='blue')
plt.xlabel('Predicted Probability')
plt.ylabel('Loss')
plt.title('BCE vs Weighted BCE')
plt.legend()
plt.grid(True)
plt.show()
```
